```python
#import dependency
import pandas as pd
from bs4 import BeautifulSoup, SoupStrainer
import requests
import numpy as np
import requests
import json
```

```python
#read csv file
books_df = pd.read_csv("bestsellers.csv")
```

```python
#counting the books written by particular author in descending order
count_authors_df = books_df.groupby("Author").count()
famous_authors_count_df = count_authors_df.sort_values("Name", ascending = False)
famous_authors_count_df.head(10)
```

```python
count_authors_df["Author"]
```

```python
# books in order of reveiw rating
ratings_df = books_df.sort_values("User Rating", ascending = False)
ratings_df.head(10)
```

```python
# authors in order of review rating
grouped_author_df = books_df.groupby("Author").mean("User Rating")
high_ratings_df = grouped_author_df.sort_values("User Rating", ascending = False)
high_ratings_df.head(10)
```

```python
#renaming the columns
renamed_df = books_df.rename(columns = {"User Rating": "Rating"})
renamed_df.to_csv("new_books.csv")
```
